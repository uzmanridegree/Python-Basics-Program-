```python
df = pd.DataFrame(employee_data)
```

```python
df
```

```python
df["Join_Date"].value_counts()
```

```python
df ["Join_Date"] = pd.to_datetime(df["Join_Date"])
```

```python
employees_after_2020 = df[df["Join_Date"] > "2020-01-01"]
print("employees who joined after January 1,2020\n", employees_after_2020[["Name","Join_Date"]])
```

```python
average_salary = df.groupby("Department")["Salary"].mean()
print("\nAverage Salary per Department:\n", average_salary)
```

```python
highest_paid_department = average_salary.max()
print("\nDepartment with the highest average salary:", highest_paid_department)
```

```python
today = datetime.today()
df["total_years_worked"] = (pd.to_datetime(today)-df["Join_Date"]).dt.days/365
```
