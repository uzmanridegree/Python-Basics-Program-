```python
import requests
```

```python
url = 'https://jsonplaceholder.typicode.com/posts'
```

```python
response = requests.get(url=url)
```

```python
response.json()
```

```python
type(response.json())
```

```python
for _ in response.json():
    print(f'post Number is {_["id"]}')
```

```python

import requests
url = 'https://api.github.com/search/repositories'
```

```python
parms = {
    'q' : 'requests+language:python',
    'sort' : 'stars',
    'order' : 'desc'
}
```

```python
response = requests.get(url = url,params=parms)
```

```python
response.json()
```
